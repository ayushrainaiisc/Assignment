```python
""" Q3. What errors are defined in the ArithmeticError class? Explain any two with an example. """

""" The ArithmeticError class in Python is a subclass of the Exception class and serves as the base class for several built-in exceptions that 
occur during arithmetic operations. """

# We will discuss ZeroDivisionError and OverFlowError:
# ZeroDivisionError

try:
  a = 10
  print(a/0)
except ZeroDivisionError as e:
  print(e)

print("ZeroDivisionError handled successfully")
print("")
# OverFlowError
try :
  a = 2
  print(float(a**1000000)) # we are trying to convert a very large integer to float
  
except OverflowError as o:
  print(o)

print("Overflow exception handled successfully")
```

```python
""" Q5). Explain ImportError. What is ModuleNotFoundError? """

""" ImportError is a built-in exception class in Python that is raised when a module or package cannot be imported. This can happen
 for a variety of reasons, such as the module not being installed, a misspelling in the module name, or a circular import """

try:
    import ayush_module
except ImportError:
    print("Failed to import module")

print("")


"""ModuleNotFoundError is a subclass of ImportError that was introduced in Python 3.6. It is raised specifically when a module cannot be 
found, and it is intended to make it clearer to developers that the issue is related to the module not being found, as opposed to a general import error."""

try:
    import ayush_module
except ModuleNotFoundError:
    print("Module not found")
```

```python
""" Q6) List down some of the best practices of exception handling """

""" Here are some best practices for exception handling in Python: """

# Use specific exception types: Try to catch specific exception types rather than catching a general Exception class. 
# This makes it easier to identify and handle different types of errors separately.

# Keep the try block small: Only include the code that may raise an exception in the try block. This makes it easier to identify
# which code may raise an exception and helps to limit the impact of the try block.

# Avoid using a bare except block: It's generally not a good idea to use a bare except block, which catches all exceptions, as it can hide 
# errors and make debugging more difficult.

# Use finally block for cleanup: Use a finally block to ensure that any necessary cleanup code is always executed, 
# regardless of whether an exception is raised.

# Reraise exceptions sparingly: Reraising an exception can be useful for logging or for passing the error up the call stack, 
# but it should be used sparingly. In most cases, it's better to handle the exception directly in the current context.

# Use custom exception types: Define custom exception types when appropriate to make it easier to identify and handle specific types of errors.

# Log exceptions: Use a logging framework to log exceptions so that you can track errors and diagnose problems.

# Keep error messages informative: When raising an exception or printing an error message, make sure the message is informative 
# and provides enough information to diagnose the error.

# Follow PEP 8 guidelines: Adhere to the PEP 8 guidelines for code style, including exception handling. This helps ensure consistency and 
# readability across different codebases.
```
